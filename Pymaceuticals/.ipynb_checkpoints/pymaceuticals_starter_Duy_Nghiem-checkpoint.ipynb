### Analysis
- In terms of Summary Statistics, Capomulin and Ramicane yielded smaller average tumor volumes than the other drugs, meanwhile Ketapril had yielded the largest average tumor volume

- The sex ratio of the mice were almost even with a 51:49 male to female ratio

- When it comes to average weight vs. average tumor volume, there is a strong and positive correlation of 0.84, however the p value for the two variables is 1.32, which is a significantly high p-value. Despite the statistically significant sample size and the strong correlation, the large p value states that we fail to reject the null hypothesis.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study_df = pd.merge(mouse_metadata, study_results, on = "Mouse ID")
mouse_study_df = mouse_study_df[["Mouse ID", "Timepoint", "Tumor Volume (mm3)", "Metastatic Sites",
                                 "Drug Regimen", "Sex", "Age_months", "Weight (g)"]]

# Display the data table for preview
mouse_study_df.head()
```

```python
# Checking the number of mice.
len(mouse_study_df["Mouse ID"].value_counts())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# get id using .loc and .unique
# subset for duplicated is Mouse ID and Timepoint, what we are looking for in .loc is unique Mouse ID
dupe_mice = mouse_study_df.loc[mouse_study_df.duplicated(subset = ["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
# output dupe_mice
dupe_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
# make dupe_mouse_df with Mouse ID of g989
dupe_mice_df = mouse_study_df[mouse_study_df["Mouse ID"] == "g989"]
dupe_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = mouse_study_df[mouse_study_df["Mouse ID"] != "g989"]
cleaned_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_df["Mouse ID"].value_counts())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

# make groupby of drug regimen from tumor volume
cleaned_group = cleaned_df["Tumor Volume (mm3)"].groupby(cleaned_df["Drug Regimen"])
# mean
mean = cleaned_group.mean()
# median
median = cleaned_group.median()
# variance
variance = cleaned_group.var()
# stdev
stdev = cleaned_group.std()
# standard error
sem = cleaned_group.sem()

# make summary stats df
stats_df = pd.DataFrame({"Mean Tumor Volume": mean,
                         "Median Tumor Volume": median,
                         "Tumor Volume Variance": variance,
                         "Tumor Volume Std. Dev.": stdev,
                         "Tumor Volume Std. Err.": sem})
stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_df = cleaned_group.agg(['mean', 'median', 'var', 'std', 'sem'])
agg_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# make dataframes
cap_df = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Capomulin"]
ram_df = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Ramicane"]
inf_df = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Infubinol"]
cef_df = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Ceftamin"]
# combine everything
drug_list = [cap_df, ram_df, inf_df, cef_df]
drugs_df = pd.concat(drug_list)
drugs_df
```

```python
# Start by getting the last (greatest) timepoint for each mouse
# groupby for Every Mouse
every_mouse_group = mouse_study_df.groupby(["Mouse ID"])
# max function
time_max = every_mouse_group["Timepoint"].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_df = pd.merge(drugs_df, time_max, on = "Mouse ID")
# get the max times
merge_df["MaxTime"] = merge_df["Timepoint_x"] == merge_df["Timepoint_y"]
# take out rows with no maxtime
merge_df.drop(merge_df.loc[merge_df["MaxTime"] == False].index, inplace = True)
merge_df.rename(columns = {"Timepoint_x":"Maximum Timepoint"}, inplace = True)

cleaned_merge_df = merge_df[["Mouse ID", "Drug Regimen", "Maximum Timepoint", "Tumor Volume (mm3)"]]
cleaned_merge_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_plots = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for i in treatments:
    # get volume
    volume = cleaned_merge_df.loc[cleaned_merge_df["Drug Regimen"] == i]["Tumor Volume (mm3)"]
    tumor_vol_plots.append(volume)
    # Locate the rows which contain mice on each drug and get the tumor volumes
    quartiles = volume.quantile([0.25, 0.50, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q-lower_q

    # Determine outliers using upper and lower bounds
    lower_bound = lower_q - (1.5*iqr)
    upper_bound = upper_q + (1.5*iqr)
    # count outliers
    outliers = volume.loc[(volume > upper_bound) | (volume < lower_bound)]
    # print
    print(f"{i}'s potential outliers: {outliers}'")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol_plots, labels = treatments, flierprops = {'marker': 'o',
                                                               'markersize': 10,
                                                               'markerfacecolor': 'red'})
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# scatter plot for weight vs tumor volume = doing a groupby
mice_group = cap_df.groupby("Mouse ID")
# make variables for new df
mean_weight = mice_group["Weight (g)"].mean()
mean_tumor = mice_group["Tumor Volume (mm3)"].mean()

scatter_df = pd.DataFrame({"Weight (g)":mean_weight,
                          "Tumor Volume (mm3)":mean_tumor})
scatter_df
```

```python
# make x and y axis
x_axis = scatter_df["Weight (g)"]
y_axis = scatter_df["Tumor Volume (mm3)"]

# plot scatter plot
plt.scatter(x_axis, y_axis, marker = 'o', facecolors = "tab:blue")
plt.xlabel("Weight")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# x and y axis just in case
x_axis = scatter_df["Weight (g)"]
y_axis = scatter_df["Tumor Volume (mm3)"]
# invoke linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_axis, y_axis)
# regression equation
regress_y = x_axis*slope + intercept
regress_eq = "y = " + str(round(slope, 2)) + "x + " +str(round(intercept, 2))
```

```python
# remake scatter plot
plt.scatter(x_axis, y_axis, marker = 'o', facecolors = "tab:blue")
plt.xlabel("Weight")
plt.ylabel("Average Tumor Volume (mm3)")

# add regression info

#regression line = plot line
plt.plot(x_axis, regress_y, color = "red")

# calculate correlation using scipy
correlation = st.pearsonr(x_axis,y_axis)
# has to be index 0 for rounding because there is also p value
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0], 2)}")
plt.show()
```
